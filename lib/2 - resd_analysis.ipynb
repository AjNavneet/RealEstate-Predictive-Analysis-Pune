```python
# Importing the required libraries
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.preprocessing import StandardScaler, LabelEncoder, OneHotEncoder

sns.set(font_scale=1.5)
```

```python
# Read the cleaned dataset
df = pd.read_csv( "../input/resd_clean.csv" )
```

#### Preliminary exploration

```python
# Check for the shape and columns
df.shape, df.columns
```

```python
 # Check for the unique values
df["company_name"].unique(), df["township"].unique()
```

```python
# Drop specified columns and remove duplicates
df = df.drop(columns=["index", "company_name", "township"]).drop_duplicates()
```

## Converting binary columns to dummy variables

```python
# Select binary columns
binary_cols = df.iloc[:, 4:].columns.to_list()
```

```python
# Calculate the average price for apartments with and without a swimming pool
df[["price", "pool"]].groupby("pool")["price"].mean()
```

```python
# Transform 'yes' and 'no' values to 1 and 0 using LabelEncoder for binary columns
laben = LabelEncoder()
for col in binary_cols:
    df[col] = laben.fit_transform(df[col])
    
# Display the last 10 rows of the DataFrame
df.tail(10)
```

---

## Sub areas contribution

```python
# Calculate the contribution of different sub-areas in the dataset
df_sa_count = df.groupby("sub_area")["price"].count().reset_index()\
                .rename(columns={"price":"count"})\
                .sort_values("count", ascending=False)\
                .reset_index(drop=True)
df_sa_count["sa_contribution"] = df_sa_count["count"]/len(df)
df_sa_count
```

```python
# Visualize the distribution of sub-areas' contribution using a bar plot
fig, ax = plt.subplots(figsize=(12, 8))
sns.barplot(data=df_sa_count, x="count", y="sub_area", ax=ax)
plt.show()
```

```python
# Calculate the number of sub-areas and the mean count of contributions
len(df_sa_count), np.mean(df_sa_count["count"])
```

```python
# Visualize price vs. surface for top 3 sub-areas
fig, ax = plt.subplots(figsize=(12, 6))

# Merge data and filter sub-areas with count > 11
df_plot = df_sa_count.merge(df[["sub_area", "surface", "price"]], on="sub_area")
df_plot = df_plot[df_plot["count"] > 11]

# Create a scatter plot
sns.scatterplot(data=df_plot, x="surface", y="price", hue="sub_area")
plt.show()
```

# Feature engineering

```python
# Select sub-areas with a count greater than 7
sa_sel_col = df_sa_count.loc[df_sa_count["count"] > 7, "sub_area"].to_list()

# Replace sub-areas not in 'sa_sel_col' with "other"
df["sub_area"] = df["sub_area"].where(df["sub_area"].isin(sa_sel_col), "other")
```

```python
# Get the unique sub-areas in the DataFrame and count them
df["sub_area"].unique(), len(df["sub_area"].unique())
```

---

#### Converting Sub-Area to dummy variables

```python
# Dummy variable conversion
hoten = OneHotEncoder(sparse=False)
X_dummy = hoten.fit_transform( df[["sub_area"]] )
X_dummy.shape, X_dummy, X_dummy.dtype
```

```python
# Type conversion
X_dummy = X_dummy.astype("int64")
```

```python
# Define column names for dummy variables
sa_cols_name = ["sa" + str(i + 1) for i in range(X_dummy.shape[1])]

# Add the dummy columns to the dataset
df.loc[:, sa_cols_name] = X_dummy

# Display the updated DataFrame
df.head()
```

```python
# Display the relationship between 'sub_area' and dummy columns
df[["sub_area"] + sa_cols_name].drop_duplicates()\
    .sort_values("sub_area").reset_index(drop=True)
```

#### Correlation map

```python
# Create a correlation heatmap for selected columns
corr_cols_list = float_cols + binary_cols
correl = data[corr_cols_list].corr()

# Generate a mask to hide the upper triangle
mask = np.zeros_like(correl)
mask[np.triu_indices_from(mask)] = True

# Plot the heatmap with annotations
fig, ax = plt.subplots(figsize=(12, 7))
sns.heatmap(correl, mask=mask, vmin=-1, cmap='Greys', annot=True, fmt='.1g')
plt.show()
```

## Feature selection

```python
# Calculate the correlation between 'price' and all other selected columns
corr_cols_list.remove("price")
corr_cols_list.extend(sa_cols_name)

corr_list = []  # To store the correlations with 'price'

# Calculate and round the correlations
for col in corr_cols_list:
    corr_list.append(round(data["price"].corr(data[col]), 2))

corr_list
```

```python
# Create a DataFrame to display sorted correlations with 'price'
df_corr = pd.DataFrame(data=zip(corr_cols_list, corr_list),
                       columns=["col_name", "corr"])\
             .sort_values("corr", ascending=False)\
             .reset_index(drop=True)
df_corr
```

```python
# Create a histogram to visualize the distribution of correlations with 'price'
fig, ax = plt.subplots(figsize=(15, 5))
abs(df_corr["corr"]).hist(bins=50)
ax.set_xticks(np.arange(0, 1, 0.05))
plt.xticks(rotation=45)
plt.show()
```

```python
# Select features with a correlation greater than 0.1 with 'price'
features = df_corr.loc[abs(df_corr["corr"]) > 0.1, "col_name"].to_list()

# Display the selected features and their count
features, len(features)
```

```python
# Create a list with sub-areas' names, mean price, and frequency (count)
sa_feature_list = [sa for sa in features if "sa" in sa]
lst = []
for col in sa_feature_list:
    sa_trigger = df[col] == 1
    sa = df.loc[sa_trigger, "sub_area"].to_list()[0]
    x = df.loc[sa_trigger, "price"]
    lst.append((sa, np.mean(x), df[col].sum()))

lst
```

```python
# Create a DataFrame for sub-area mean price comparison
sa_f = pd.DataFrame(lst, columns=["sub_area", "mean_price", "count"])
sa_f["total_mean_price"] = np.mean(df["price"])
sa_f
```

## Data scalling

```python
# Select the final dataset with the chosen features and the 'price' column
sel_data = data[features + ["price"]].copy()
sel_data.head()
```
